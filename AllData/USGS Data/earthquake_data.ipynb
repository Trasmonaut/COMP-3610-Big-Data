```python
import pandas as pd

df = pd.read_csv('earthquake data 2000 - 2025.csv', header=0)
```

```python
df
```

```python
#dropping unecessary columns
df_cleaned = df.drop(columns=['dmin', 'magNst', 'net', 'rms', 'nst', 'status', 'type'])

df_cleaned['horizontalError'].fillna(0, inplace=True)
df_cleaned['depthError'].fillna(0, inplace=True)
df_cleaned['magError'].fillna(0, inplace=True)

df_cleaned.to_csv('cleaned_earthquake_data.csv', index=False)
```
